```python
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score

range_n_clusters = np.arange(2,11,1)
print(range_n_clusters)

scalers =StandardScaler()
cards2=cards.drop(['card_pan_id','max_amount', 'sum_amount','count_amount','nb_countries','count_amount_log10'], axis=1)
cards3 = scaler.fit_transform(cards2)
card3=pd.DataFrame(cards3)
card3.columns=['recency','ecom_rate','bel_rate','monetarylog10']
#print(card3.head())
ssd=[]
silhouettes=[]
for n_clusters in range_n_clusters: 
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    kmeans.fit(cards3)
    y_pred = kmeans.predict(cards3)
    
    #cards3["label"]=y_pred
    pred = pd.DataFrame(y_pred)
    print(pred[0].value_counts())
    ssd.append(kmeans.inertia_)    
    score = silhouette_score(cards3,y_pred, sample_size=5000)
    silhouettes.append(score)
    if(n_clusters>3 and score>max_silouhette):
        max_silouhette=score
        silouhette_config=''+str(n_clusters)+' StandardScaler'
    print('--', n_clusters, score)
print('-----------------')

#plt.plot(range_n_clusters, silhouettes)
```

```python
#plt.plot(range_n_clusters, silhouettes)

fig, ax = plt.subplots() # Create the figure and axes object
# Plot the first x and y axes:
ax.plot(range_n_clusters, ssd) 
ax.set_ylabel("ssd",color="b")
ax2 = ax.twinx()

ax2.plot(range_n_clusters, silhouettes, color='r') 
ax2.set_ylabel("silhouette",color="r")
```
